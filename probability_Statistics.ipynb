```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = 5
std = 1
x = np.random.randn(1000)*std + mean
```

```python
plt.hist(x,10)
plt.show()
```

```python
np.unique(np.round(x),return_counts=True)
```

```python
mean_h = 75
mean_w = 72
std_h = 20
std_w = 21
h = np.random.randn(1000)*std_h + mean_h
w = np.random.randn(1000)*std_w + mean_w
plt.xlim(0, 200)  # Set x-axis range (height)
plt.ylim(0, 200)  # Set y-axis range (weight)
plt.scatter(h,w)
plt.show()
```

```python
#multivatiate notmal distribution 2d
mean = np.array([0.0,0.0])
cov = np.array([[1,0.8],[0.8,1]])
dist = np.random.multivariate_normal(mean,cov,500)
```

```python
plt.xlim(-20, 20)  # Set x-axis range (height)
plt.ylim(-20, 20)  # Set y-axis range (weight)
plt.scatter(dist[:,0],dist[:,1])
```
